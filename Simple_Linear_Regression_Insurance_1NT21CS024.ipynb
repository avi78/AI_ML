## Simple Linear Regression for insurance data with Age vs Premium amt. that users pay
---



- Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

- Importing the dataset

```python
dataset = pd.read_csv('insurance_data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

- Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/5, random_state = 0)
```

- Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

- Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
```

- Visualising the Training set results

```python
plt.scatter(X_train, y_train, color = 'orange')
plt.plot(X_train, regressor.predict(X_train), color = 'green')
plt.title('Insurance Premium vs Age (Training set)')
plt.xlabel('Age')
plt.ylabel('Premium')
plt.show()
```

- Visualising the Test set results

```python
plt.scatter(X_test, y_test, color = 'orange')
plt.plot(X_train, regressor.predict(X_train), color = 'green')
plt.title('Insurance Premium vs Age (Test set)')
plt.xlabel('Age')
plt.ylabel('Premium')
plt.show()
```
